```python
import warnings
warnings.filterwarnings("ignore") 

import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
import os, librosa
from tqdm import tqdm
import torchaudio
import IPython.display as ipd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
audio_path = 'UrbanSound8K/fold2/77751-4-3-0.wav'
```

```python
import IPython.display as ipd

# Load audio
y, sr = librosa.load(audio_path, sr=16000)
ipd.Audio(y, rate=sr)
```

```python
# visualize the audio
plt.figure(figsize=(12, 4))
plt.plot(y)
plt.title('Waveform')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()
```

```python
s = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
s = librosa.power_to_db(s, ref=np.max) #converting your mel spectrogram from raw power values to decibels (dB) 
print(s.shape)
print(len(s[0])) # 8 time frames 
```

```python
## visualise the mel spectrogram
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
librosa.display.specshow(s, sr=sr, x_axis='time', y_axis='hz', fmax=8000)
plt.colorbar(format='%+2.0f dB') #to show the color bar
plt.title('Mel spectrogram')
plt.tight_layout()
plt.show()


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 5))
plt.imshow(s, aspect='auto', origin='lower', cmap='magma')
plt.colorbar(label='dB')
plt.title('Mel Spectrogram (as matrix)')
plt.xlabel('Time frame')
plt.ylabel('Mel frequency bin')
plt.tight_layout()
plt.show()
```
